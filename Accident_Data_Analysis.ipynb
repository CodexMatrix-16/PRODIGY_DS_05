# Droping columns with high missing values

```python
df.drop(columns=['End_Lat','End_Lng','Wind_Chill(F)','Precipitation(in)'],axis=1,inplace=True)
```

```python
categorical=[i for i in df.columns if df[i].dtype=='O']
#for categorical values we can replace the null values with the Mode of it
for i in categorical:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

```python
df.drop(columns=['Wind_Speed(mph)', 'Visibility(mi)', 'Pressure(in)', 'Humidity(%)', 'Temperature(F)'], axis=1, inplace=True)
```

# Exploratory Data Analysis

```python
city_accidents = df['City'].value_counts().sort_values(ascending = False).reset_index()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data= city_accidents.head(10),x='City',y='count',palette='viridis')
plt.title("Top 10 cities with most number of accidents")
plt.ylabel("No of accidents")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data= city_accidents.tail(10),x='City',y='count',palette='coolwarm')
plt.title("Bottom 10 cities with least number of accidents")
plt.ylabel("No of accidents")
plt.xticks(rotation=45)
plt.show()
```

```python
state_accidents = df['State'].value_counts().sort_values(ascending = False).reset_index()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data= state_accidents.head(5),x='State',y='count',palette='colorblind')
plt.title("Top 5 States with most number of accidents")
plt.ylabel("No of accidents")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data= state_accidents.tail(5),x='State',y='count',palette='colorblind')
plt.title("Bottom 5 States with least number of accidents")
plt.ylabel("No of accidents")
plt.show()
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'], errors='coerce')

# Extract the year from the 'Start_Time' column and store it in a new column called 'Year'
df['Year'] = df['Start_Time'].dt.year

print(df[['Start_Time', 'Year']].head())
```

```python
year_count = df['Year'].value_counts().reset_index()
```

```python
sns.barplot(x='Year',y='count',data=year_count,palette='viridis')
plt.title("Years with most number of accidents")
plt.show()
```

```python
plt.figure(figsize=(14,10))
sns.scatterplot(x=df['Start_Lng'],y=df['Start_Lat'],hue=df['State'])
plt.legend(loc="lower right")
plt.show()
```

```python
weather = df['Weather_Condition'].value_counts().sort_values(ascending=False).reset_index()
sns.barplot(x='count',y='Weather_Condition',data=weather[:10],orient='horizontal',palette='viridis')
plt.show()
```

# Conclusion

**In this traffic data analysis project, we undertook a comprehensive examination of the accident records, including data cleaning, exploratory data analysis (EDA), and visualization. Here are the key findings and observations from the analysis:**

-  1) Miami stands out as the city with the highest number of accidents. This could be attributed to its high population density and traffic volume.
-  2) On the other hand, Star Junction and Stromsburg are among the cities with the least number of accidents, possibly due to their smaller size and lower traffic.
-  3) California reports the highest number of accident cases, which aligns with its large population and extensive road network.
-  4) Conversely, South Dakota (SD) has the least number of accident cases, likely due to its lower population density and traffic volume.
-  5) The year 2021 recorded the highest accident rates, indicating a potential increase in traffic or changes in reporting during this period
-  6) Fair weather conditions were prevalent during most of the accidents, indicating that factors other than adverse weather, such as driver behavior or traffic congestion, may play a significant role in accident occurrences.

**These observations provide valuable insights into traffic patterns and accident hotspots, which can inform policy decisions, urban planning, and safety measures aimed at reducing the incidence and severity of accidents.**
